```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron

# Datos (igual que antes)
data = {
    'PBI': [2.7, 0.6, 5.5, 4.2, 5, 6.3, 7.5, 8.5, 9.1, 1.1,
            8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
            -10.9, 13.4, 2.8, -0.4, 3.3],
    'Gasto_publico': [19, 18.2, 17.5, 17.5, 17.1, 17.7, 17, 17.5, 17.1, 18.2,
                      17.9, 17.8, 17.9, 18.5, 19.4, 19.2, 17.3, 17.9, 17.9, 18.4,
                      23.7, 21.1, 21.0, 19.9, 19.7],
    'Inflacion': [3.8, 2.0, 0.2, 2.3, 3.7, 1.6, 2.0, 1.8, 5.8, 2.9,
                  1.5, 3.37, 3.61, 2.77, 3.41, 3.40, 3.56, 2.99, 1.51, 2.25,
                  2.00, 4.27, 8.33, 6.46, 2.01]
}

df = pd.DataFrame(data)
df['Clase'] = (df['PBI'] >= 4.0).astype(int)

X = df[['Gasto_publico', 'Inflacion']]
y = df['Clase']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

modelo = Perceptron(tol=1e-3, max_iter=1000, eta0=0.01, random_state=42)
modelo.fit(X_scaled, y)

pesos = modelo.coef_
sesgo = modelo.intercept_

# Función escalón (activación)
def funcion_escalon(valor):
    return 1 if valor >= 0 else 0

# Función para predecir manualmente dado gasto e inflación
def predecir_pbi(gasto, inflacion):
    entrada = np.array([[gasto, inflacion]])
    entrada_escalada = scaler.transform(entrada)
    z = np.dot(entrada_escalada, pesos.T) + sesgo
    return funcion_escalon(z[0][0])

# Ejemplos:
gasto_ejemplo = 20
inflacion_ejemplo = 2.5

resultado = predecir_pbi(gasto_ejemplo, inflacion_ejemplo)
print(f"Para Gasto público={gasto_ejemplo} e Inflación={inflacion_ejemplo}:")
print("Predicción del PBI alto (1) o bajo (0):", resultado)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron

# Datos reales
data = {
    'PBI': [2.7, 0.6, 5.5, 4.2, 5, 6.3, 7.5, 8.5, 9.1, 1.1,
            8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
            -10.9, 13.4, 2.8, -0.4, 3.3],
    'Gasto_publico': [19, 18.2, 17.5, 17.5, 17.1, 17.7, 17, 17.5, 17.1, 18.2,
                      17.9, 17.8, 17.9, 18.5, 19.4, 19.2, 17.3, 17.9, 17.9, 18.4,
                      23.7, 21.1, 21.0, 19.9, 19.7],
    'Inflacion': [3.8, 2.0, 0.2, 2.3, 3.7, 1.6, 2.0, 1.8, 5.8, 2.9,
                  1.5, 3.37, 3.61, 2.77, 3.41, 3.40, 3.56, 2.99, 1.51, 2.25,
                  2.00, 4.27, 8.33, 6.46, 2.01]
}

df = pd.DataFrame(data)
df['Clase'] = (df['PBI'] >= 4.0).astype(int)

X = df[['Gasto_publico', 'Inflacion']]
y = df['Clase']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

modelo = Perceptron(tol=1e-3, max_iter=1000, eta0=0.01, random_state=42)
modelo.fit(X_scaled, y)

pesos = modelo.coef_
sesgo = modelo.intercept_

print("Pesos aprendidos:", pesos)
print("Sesgo:", sesgo)

# Función escalón (activación)
def funcion_escalon(valor):
    return 1 if valor >= 0 else 0

# Predicciones manuales usando pesos, sesgo y función escalón
predicciones = []
for x in X_scaled:
    z = np.dot(x, pesos.T) + sesgo
    prediccion = funcion_escalon(z[0])
    predicciones.append(prediccion)

print("Predicciones:", predicciones)
print("Clase real:  ", list(y))
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, predicciones)
print(f"Exactitud del modelo: {accuracy:.2f}")

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

# Datos económicos del Perú (2000-2024)
data = {
    'PBI': [2.7, 0.6, 5.5, 4.2, 5, 6.3, 7.5, 8.5, 9.1, 1.1,
            8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
            -10.9, 13.4, 2.8, -0.4, 3.3],
    'Gasto_publico': [19, 18.2, 17.5, 17.5, 17.1, 17.7, 17, 17.5, 17.1, 18.2,
                      17.9, 17.8, 17.9, 18.5, 19.4, 19.2, 17.3, 17.9, 17.9, 18.4,
                      23.7, 21.1, 21.0, 19.9, 19.7],
    'Inflacion': [3.8, 2.0, 0.2, 2.3, 3.7, 1.6, 2.0, 1.8, 5.8, 2.9,
                  1.5, 3.37, 3.61, 2.77, 3.41, 3.40, 3.56, 2.99, 1.51, 2.25,
                  2.00, 4.27, 8.33, 6.46, 2.01]
}
df = pd.DataFrame(data)

# Crear variable de clase (1 si PBI >= 4%)
df['Clase'] = (df['PBI'] >= 4.0).astype(int)

# Variables independientes (entradas)
X = df[['Gasto_publico', 'Inflacion']]
y = df['Clase']

# Escalar variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Entrenar perceptrón con más iteraciones y tasa de aprendizaje ajustada
modelo = Perceptron(tol=1e-3, max_iter=5000, eta0=0.001, random_state=42)
modelo.fit(X_scaled, y)

# Pesos y sesgo aprendidos
pesos = modelo.coef_
sesgo = modelo.intercept_

print("Pesos aprendidos:", pesos)
print("Sesgo:", sesgo)

# Predicciones con el modelo entrenado
predicciones = modelo.predict(X_scaled)

# Mostrar predicciones y clases reales
print("Predicciones:", list(predicciones))
print("Clase real: ", list(y))

# Calcular y mostrar la exactitud
accuracy = accuracy_score(y, predicciones)
print(f"Exactitud del modelo: {accuracy:.2f}")
```

```python
import math
import turtle

turtle.bgcolor("black")
turtle.shape("turtle")
turtle.speed(300)

turtle.goto(0, -40)
turtle.pendown()
h = 0
phi = 137.508 * (math.pi / 180.0)
for i in range(16):
    for j in range(18):
        turtle.color("yellow")
        h += 0.005
        turtle.rt(90)
        turtle.circle(150 - j * 6, 90)
        turtle.lt(90)
        turtle.circle(150 - j * 6, 90)
        turtle.rt(180)
    turtle.circle(40, 24)

turtle.penup()
turtle.goto(0, 0)
turtle.color("black")
turtle.fillcolor("brown")
turtle.begin_fill()
turtle.circle(0)
turtle.end_fill()

phi = 137.508 * (math.pi / 180.0)
for i in range(160 + 40):
    r = 4 * math.sqrt(i)
    theta = i * phi
    x = r * math.cos(theta)
    y = r * math.sin(theta)
    turtle.penup()
    turtle.goto(x, y)
    turtle.setheading(i * 137.508)
    turtle.pendown()
    if i < 160:
        turtle.stamp()

turtle.penup()
turtle.goto(0, 300)
turtle.color("red")
turtle.write(":)", align="center", font=("Arial", 50, "bold"))


turtle.Screen().exitonclick()
```
