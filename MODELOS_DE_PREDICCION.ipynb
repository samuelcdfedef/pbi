Se buscara estimar el crecmiento del PBI con las 5 variables explicativas: consumo, inversion, gasto publico, XN y tasa de inflacion

 Arquitectura del modelo: Se utilizó una red neuronal con tres capas: una capa de entrada con 5 neuronas (inflacion, gasto público, XN, consumo privado, inversiones ), una capa oculta de 10 neuronas con activación ReLU, y una capa de salida con una neurona y activación lineal. 3.1.1 PREPARACION DE LOS DATOS y RESULTADOS

```python
# 2. Preparar entradas y salidas
X = df[['Gasto_publico', 'Inflacion', 'Inversion', 'Consumo', 'XN']].values
y = df['PBI'].values

# 3. Normalizar entradas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5. Construir el modelo
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(5,)),
    keras.layers.Dense(1)
])

# 6. Compilar
model.compile(optimizer='adam', loss='mse')

# 7. Entrenar con EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=200, batch_size=2, validation_data=(X_test, y_test),
                    callbacks=[early_stop], verbose=1)

# 8. Evaluar el modelo
test_loss = model.evaluate(X_test, y_test)
print(f"\nError en datos de prueba (MSE): {test_loss:.2f}")
print(f"Raíz del error (RMSE): {np.sqrt(test_loss):.2f}")

```

Después de entrenar el modelo durante 200 iteraciones, obtuvimos un error cuadrático medio (MSE) de 7.38 y una raíz del error (RMSE) de 2.72. Esto significa que, en promedio, nuestras predicciones se desviaron alrededor de 2.72 puntos porcentuales del valor real del PBI.

Si bien el modelo no captura con exactitud eventos extremos, como la fuerte caída del 2020 por la pandemia, sí logra reflejar bien la tendencia general del crecimiento económico. Esto muestra que, incluso con una arquitectura simple, las redes neuronales con activación ReLU pueden ser una herramienta útil

3.2 MODELO ARIMA

Se buscara modelar  el crecimiento del PBI real en Perú utilizando únicamente la serie histórica del propio PBI (% anual) mediante un modelo ARIMA

Debemos tener presente que:

-Se utilizó la serie anual de crecimiento del PBI (2000–2024).

-Se aplicó la prueba ADF y análisis visual (ACF y PACF) para evaluar estacionariedad.

-Se realizó diferenciación de primer orden, necesaria por no estacionariedad inicial.

-Búsqueda automática del mejor modelo ARIMA (p, d, q) según el criterio AIC.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")

# 1. Datos: PBI real anual (%)
pbi = [
    2.7, 0.6, 5.5, 4.2, 5, 6.3, 7.5, 8.5, 9.1, 1.1,
    8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
    -10.9, 13.4, 2.8, -0.4, 3.3
]
años = list(range(2000, 2025))
df = pd.DataFrame({'Año': años, 'PBI': pbi})
df.set_index('Año', inplace=True)

# 2. Graficar serie original
plt.figure(figsize=(10, 4))
plt.plot(df['PBI'], marker='o', color='darkblue')
plt.title("Serie original: PBI (%) Perú (2000–2024)")
plt.xlabel("Año")
plt.ylabel("PBI (%)")
plt.grid(True)
plt.show()

# 3. Diferenciación de primer orden
df['PBI_diff'] = df['PBI'].diff()
df.dropna(inplace=True)

plt.figure(figsize=(10, 4))
plt.plot(df['PBI_diff'], marker='o', color='purple')
plt.title("Serie estacionaria (1ra diferencia del PBI)")
plt.xlabel("Año")
plt.ylabel("ΔPBI (%)")
plt.grid(True)
plt.show()

# 4. Correlogramas ACF y PACF
plot_acf(df['PBI_diff'], lags=10)
plt.title("ACF de la serie diferenciada")
plt.show()

plot_pacf(df['PBI_diff'], lags=10)
plt.title("PACF de la serie diferenciada")
plt.show()

# 5. Búsqueda del mejor ARIMA(p,1,q)
mejor_aic = np.inf
mejor_orden = None
mejor_modelo = None

for p in range(0, 4):
    for d in range(1, 2):  # fija en 1 por diferenciación
        for q in range(0, 4):
            try:
                modelo = ARIMA(df['PBI'], order=(p, d, q)).fit()
                aic = modelo.aic
                if aic < mejor_aic:
                    mejor_aic = aic
                    mejor_orden = (p, d, q)
                    mejor_modelo = modelo
            except:
                continue

```

3.2.1 IDENTIFICACION DEL MODELO Y ESTIMACION DEL MODELO


PARA ACF:
 los primeros dos rezagos son significativos, y después desaparece la autocorrelación.

 PARA PACF: Un pico grande en el rezago 1, luego todos los siguientes valores están dentro del intervalo, es decir, no significativo. Esto sugiere que no hay estructura autorregresiva fuerte.

El ARIMA SERA ARIMA(0,1,2)

3.2.2 DIAGNOSTICO DEL MODELO Y EVALUACION DE RESULTADOS



```python
print(f"Mejor modelo ARIMA según AIC: {mejor_orden} con AIC = {mejor_aic:.2f}")
print(mejor_modelo.summary())

# 6. Predicción dentro de muestra
df['PBI_predicho'] = mejor_modelo.fittedvalues

# 7. Métricas de evaluación
mse = mean_squared_error(df['PBI'], df['PBI_predicho'])
r2 = r2_score(df['PBI'], df['PBI_predicho'])
print(f"\nMSE (Error Cuadrático Medio): {mse:.4f}")
print(f"R² (Coef. de determinación): {r2:.4f}")



```

El modelo ARIMA(0,1,2) presentó un Error Cuadrático Medio (MSE) de 18.03, lo que indica un nivel moderado de precisión en la estimación del crecimiento del PBI. Sin embargo, el R² obtenido fue bajo (0.055), lo que refleja una capacidad limitada del modelo para explicar la variabilidad total de la serie. A pesar de ello, los residuos muestran ausencia de autocorrelación, por lo que el modelo sigue siendo estadísticamente válido como punto de comparación.

3.2.3 Visualizacion y proyeccion para 2025

```python
# 8. Predicción para 2025
pred_2025 = mejor_modelo.forecast(steps=1)
pbi_2025 = pred_2025.iloc[0]
print(f"\nPredicción del PBI para 2025: {pbi_2025:.2f}%")

# 9. Gráfico comparativo con proyección
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['PBI'], label='PBI Real', marker='o')
plt.plot(df.index, df['PBI_predicho'], label=f'Estimado ARIMA{mejor_orden}', linestyle='--', marker='s')
plt.axvline(x=2024, color='gray', linestyle='--')
plt.plot(2025, pbi_2025, 'ro', label='Proyección 2025')
plt.title("PBI Real, Estimado y Proyección 2025 (ARIMA)")
plt.xlabel("Año")
plt.ylabel("PBI (%)")
plt.grid(True)
plt.legend()
plt.show()
```

 3.3 MODELO ARIMAX

Para este modelo utilizaremos un Arima sumado con una variable exogena. Se incorpora una variable exógena (dummy = 1 en 2020). De esta manera se podra captar el efecto del covid en el PBI

Yt= ARIMA(p,d,q)+wDt+etYt

*W: mide el impacto directo del evento

*D= 1 en 2020 (pulso), o de 2020 en adelante (escalón)






```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, r2_score

# 1. Base de datos
pbi = [
    2.7, 0.6, 5.5, 4.2, 5, 6.3, 7.5, 8.5, 9.1, 1.1,
    8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
    -10.9, 13.4, 2.8, -0.4, 3.3
]
años = list(range(2000, 2025))
df = pd.DataFrame({'Año': años, 'PBI': pbi})
df.set_index('Año', inplace=True)

# 2. Dummy COVID
df['dummy_covid'] = (df.index == 2020).astype(int)

# 3. Ajuste modelo ARIMAX(0,1,2)
modelo_arimax = ARIMA(df['PBI'], order=(0, 1, 2), exog=df[['dummy_covid']])
res_arimax = modelo_arimax.fit()

# 4. Evaluación
df['PBI_predicho_ARIMAX'] = res_arimax.fittedvalues
mse = mean_squared_error(df['PBI'], df['PBI_predicho_ARIMAX'])
r2 = r2_score(df['PBI'], df['PBI_predicho_ARIMAX'])

print(res_arimax.summary())
print(f"\nMSE: {mse:.4f}")
print(f"R²: {r2:.4f}")
```

3.3.1 Grafica de resultados y proyeccion para 2025

```python
# 5. Proyección para 2025

exog_2025 = pd.DataFrame({'dummy_covid': [0]})
forecast_2025 = res_arimax.get_forecast(steps=1, exog=exog_2025)

# Obtener valores proyectados y de intervalo
pbi_2025 = forecast_2025.predicted_mean.values[0]
conf_int_2025 = forecast_2025.conf_int(alpha=0.05).values[0]

print(f"\nProyección del PBI para 2025: {pbi_2025:.2f}%")
print(f"Intervalo de confianza al 95%: [{conf_int_2025[0]:.2f}%, {conf_int_2025[1]:.2f}%]")

# 6. Gráfico incluyendo la proyección
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['PBI'], label='PBI Real', marker='o')
plt.plot(df.index, df['PBI_predicho_ARIMAX'], label='Estimado ARIMAX', linestyle='--', marker='s')
plt.axvline(x=2024, color='gray', linestyle='--')
plt.plot(2025, pbi_2025, 'ro', label='Proyección 2025')
plt.fill_between([2025], conf_int_2025[0], conf_int_2025[1], color='pink', alpha=0.5)
plt.title("PBI Real, Estimado y Proyección 2025 (ARIMAX)")
plt.xlabel("Año")
plt.ylabel("PBI (%)")
plt.grid(True)
plt.legend()
plt.show()
```

El objetivo sera aplicar el filtro de Kalman a la serie del PBI de Perú (2000–2024) para estimar el valor “real” libre de ruido, y proyectar el crecimiento para el año 2025.



*Predicción del estado
Se usó el método kf.filter() para estimar el estado verdadero del PBI en cada año. Luego, se aplicó kf.filter_update() para predecir el valor de 2025, usando como base el último estado estimado (2024) y su covarianza.


*Ganancia de Kalman (implícita)
La ganancia de Kalman se calcula automáticamente en el filtro. Este valor determina el peso entre la predicción previa y la nueva observación.
Aunque no se imprime directamente, es parte fundamental del proceso de estimación y se utiliza internamente en el algoritmo de pykalman.

*Actualización del estado
Cuando se cuenta con observación (años 2000–2024), el filtro actualiza el estado estimado. En el caso del año 2025, como no hay observación, se mantiene la predicción.



```python
# 1. Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pykalman import KalmanFilter
from sklearn.metrics import mean_squared_error, r2_score

# 2. Datos: Crecimiento del PBI real de Perú 2000–2024
años = np.arange(2000, 2025)
pbi = np.array([
    2.7, 0.6, 5.5, 4.2, 5.0, 6.3, 7.5, 8.5, 9.1, 1.1,
    8.3, 6.3, 6.1, 5.9, 2.4, 3.3, 4.0, 2.5, 4.0, 2.2,
    -10.9, 13.4, 2.8, -0.4, 3.3
])

# 3. Filtro de Kalman
kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
kf = kf.em(pbi, n_iter=10)
filtered_state_means, filtered_state_covariances = kf.filter(pbi)

# 4. Proyección para el año 2025
next_state_mean, next_state_cov = kf.filter_update(
    filtered_state_means[-1],
    filtered_state_covariances[-1],
    observation=None
)
```

 3.4.1 Matrices estimadas del modelo

```python
print("Matriz de transición (A):")
print(kf.transition_matrices)

print("\nMatriz de observación (H):")
print(kf.observation_matrices)

print("\nCovarianza de transición (Q):")
print(kf.transition_covariance)

print("\nCovarianza de observación (R):")
print(kf.observation_covariance)
```

3.4.2 Visualizacion y proyeccion para 2025

```python
# Añadir la proyección a los datos
años_proyectado = np.append(años, 2025)
pbi_proyectado = np.append(pbi, np.nan)
kalman_proyectado = np.append(filtered_state_means, next_state_mean)

# 5. Métricas de ajuste (solo datos reales)
mse = mean_squared_error(pbi, filtered_state_means)
r2 = r2_score(pbi, filtered_state_means)
print(f"MSE (Error Cuadrático Medio): {mse:.4f}")
print(f"R² (Coef. de determinación): {r2:.4f}")
print(f"\n Proyección del PBI para 2025 con Kalman: {next_state_mean[0]:.2f} %")

# 6. Visualización
plt.figure(figsize=(10, 5))
plt.plot(años, pbi, 'o-', label="PBI Real")
plt.plot(años, filtered_state_means, 'r--', label="Kalman Estimado")
plt.plot(2025, next_state_mean[0], 'go', label="Proyección 2025")
plt.axvline(x=2024.5, color='gray', linestyle='--', alpha=0.5)
plt.title("Filtro de Kalman aplicado al PBI + Proyección 2025")
plt.xlabel("Año")
plt.ylabel("Crecimiento del PBI (%)")
plt.legend()
plt.grid(True)
plt.show()
```
